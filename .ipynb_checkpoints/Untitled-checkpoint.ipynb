```python
a = []
for i in range(len(data['innings'])):
    inning_number = data['innings'][i]
    team = inning_number['team']
    
    for over in inning_number['overs']:
        over_number = str(over['over'])
        for j, delivery in enumerate(over['deliveries']):
            if 'review' in delivery:
                if delivery['review']['by'] == team and delivery['review']['decision'] == 'struck down':
                    a.append({'Event': data['info']['event']['name'],
                              
                              'Innings': i+1,
                              'Team': team,
                              'Over': f"{over_number}.{j+1}",
                              'Bowler': delivery['bowler'],
                              'Batter': delivery['batter'],
                              'ReviewBy': 'BattingTeam',
                              'Umpire_Decision': delivery['review']['decision'],
                              'Is_Overturned': 0
                             })
                elif delivery['review']['by'] == team and delivery['review']['decision'] == 'upheld':
                    a.append({'Event': data['info']['event']['name'],
                              
                              'Innings': i+1,
                              'Team': team,
                              'Over': f"{over_number}.{j+1}",
                              'Bowler': delivery['bowler'],
                              'Batter': delivery['batter'],
                              'ReviewBy': 'BattingTeam',
                              'Umpire_Decision': delivery['review']['decision'],
                              'Is_Overturned': 1
                             })
                elif delivery['review']['by'] != team and delivery['review']['decision'] == 'upheld':
                    a.append({'Event': data['info']['event']['name'],
                              
                              'Innings': i+1,
                              'Team': team,
                              'Over': f"{over_number}.{j+1}",
                              'Bowler': delivery['bowler'],
                              'Batter': delivery['batter'],
                              'ReviewBy': 'BowlingTeam',
                              'Umpire_Decision': delivery['review']['decision'],
                              'Is_Overturned': 1
                             })
                else:
                    a.append({'Event': data['info']['event']['name'],
                              
                              'Innings': i+1,
                              'Team': team,
                              'Over': f"{over_number}.{j+1}",
                              'Bowler': delivery['bowler'],
                              'Batter': delivery['batter'],
                              'ReviewBy': 'BowlingTeam',
                              'Umpire_Decision': delivery['review']['decision'],
                              'Is_Overturned': 0
                             })
                
```
